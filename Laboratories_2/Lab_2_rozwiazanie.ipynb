Wykorzystaj posiadane pliki bądź użyj nowe. Użyj Sparka do odczytania jednego pliku i użyj wszystkich typów `read modes`. Poprawny plik nie wywoła żadnych efektów, więc popsuj dane tak, aby każda z `read modes` zadziałała.


```
# Przykładowe dane pogodowe z błędami
data = """date,temperature,condition
2024-04-01,13,Sunny
2024-04-02,invalid,Sunny
2024-04-03,28,
2024-04-04,,Rain
2024-04-05,17,Cloudy
2024-04-06,-1,Storm
2024-04-07,six degrees,Sunny
"""

# zapis danych
dbutils.fs.put("/shared_file/weather.csv", data)
```

```
display(dbutils.fs.ls("/shared_file/"))
```

```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType

file_path_3 = "/shared_file/weather.csv"

schema_3 = StructType([
    StructField("date", DateType(), True),   
    StructField("temperature", IntegerType(), True),  
    StructField("condition", StringType(), True) 
])
```

```
df_permissive = spark.read \
    .format("csv") \
    .option("header", "true") \
    .schema(schema_3) \
    .option("mode", "PERMISSIVE") \
    .load(file_path_3)

print("Tryb PERMISSIVE:")
display(df_permissive)
```

```
df_dropmalformed = spark.read \
    .format("csv") \
    .option("header", "true") \
    .schema(schema_3) \
    .option("mode", "DROPMALFORMED") \
    .load(file_path_3)

print("Tryb DROPMALFORMED:")
display(df_dropmalformed)
```

```
df_failfast = spark.read \
    .format("csv") \
    .option("header", "true") \
    .schema(schema_3) \
    .option("mode", "FAILFAST") \
    .load(file_path_3)

print("Tryb FAILFAST:")
display(df_failfast )
```

# Zadanie 4

`Użycie DataFrameWriter`

Zapisz jeden z wybranych plików do formatów `(‘.parquet’, ‘.json’)`. Sprawdź, czy dane są zapisane poprawnie, użyj do tego `spark.read` (DataFrameReader).

```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema_4 = StructType([
    StructField("date", StringType(), True),          
    StructField("temperature", IntegerType(), True),  
    StructField("condition", StringType(), True)    
])
```

```
file_path_4 = "/shared_file/weather.csv"

df = spark.read \
    .schema(schema_4) \
    .option("header", "true") \
    .format("csv") \
    .load(file_path_4)

print("Plik CSV:")
display(df)
```

```
parquet_path = "dbfs:/shared_file/weather.parquet"

df.write \
    .format("parquet") \
    .mode("overwrite") \
    .save(parquet_path)

df_parquet = spark.read \
    .format("parquet") \
    .load(parquet_path)

print("Plik Parquet:")
display(df_parquet)
```

```
json_path = "dbfs:/shared_file/weather.json"

df.write \
    .format("json") \
    .mode("overwrite") \
    .save(json_path)


df_json = spark.read \
    .schema(schema_4) \
    .format("json") \
    .load(json_path)

print("Plik JSON:")
display(df_json)
```
